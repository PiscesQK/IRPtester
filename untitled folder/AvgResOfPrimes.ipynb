```python
#Prime Numbers <//> Natural Numbers
import matplotlib.pyplot as plt
import numpy as np

# Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Function to generate the first n prime numbers
def generate_primes(n):
    primes = []
    i = 2
    while len(primes) < n:
        if is_prime(i):
            primes.append(i)
        i += 1
    return primes

# Function to calculate the average of residues
def average_residues(prime):
    residues = [i % prime for i in range(1, prime)]
    return np.mean(residues)

# Generate the first 1000 prime numbers
primes = generate_primes(10)

# Calculate the average of residues for each prime number
averages = [average_residues(p) for p in primes]

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(primes, averages, marker='o')
plt.xlabel('Prime Numbers')
plt.ylabel('Average of Residues')
plt.title('Average of Residues for the First 1000 Prime Numbers')
plt.grid(True)
plt.show()
```

```python
#Natural Numbers <//> Natural Numbers
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate the average of residues for a given number
def average_residues(number):
    if number > 1:
        residues = [i % number for i in range(1, number)]
        return np.mean(residues)/number
    else:
        return 0  # For number 1, there are no residues to calculate

# Set the limit for the numbers (inclusive)
limit = 10

# Calculate the average of residues for each number
averages = [average_residues(n) for n in range(1, limit + 1)]

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(range(1, limit + 1), averages, marker='o')
plt.xlabel('Numbers')
plt.ylabel('Average of Residues')
plt.title('Average of Residues for Numbers up to ' + str(limit))
plt.grid(True)
plt.show()
```

```python
#Natural Numbers <//> Prime Numbers
import matplotlib.pyplot as plt
import numpy as np

# Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Function to generate prime numbers up to a certain limit
def generate_primes(limit):
    primes = []
    for i in range(2, limit+1):
        if is_prime(i):
            primes.append(i)
    return primes

# Function to calculate the average of residues for a number using primes
def average_residues_with_primes(number, primes):
    relevant_primes = [p for p in primes if p <= number]
    if relevant_primes:
        residues = [number % p for p in relevant_primes]
        return np.mean(residues)/number
    else:
        return 0

# Set the limit for the numbers (inclusive)
limit = 500000

# Generate prime numbers up to the limit
primes = generate_primes(limit)

# Calculate the average of residues for each number using primes
averages = [average_residues_with_primes(n, primes) for n in range(1, limit + 1)]

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(range(1, limit + 1), averages, marker='o')
plt.xlabel('Numbers')
plt.ylabel('Average of Residues with Primes')
plt.title('Average of Residues with Primes for Numbers up to ' + str(limit))
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Function to generate prime numbers up to a certain limit
def generate_primes(limit):
    primes = []
    for i in range(2, limit+1):
        if is_prime(i):
            primes.append(i)
    return primes

# Function to calculate the average of residues for a prime number using smaller primes
def average_residues_prime_vs_prime(prime, smaller_primes):
    if smaller_primes:
        residues = [prime % p for p in smaller_primes]
        return np.mean(residues)
    else:
        return 0

# Generate prime numbers up to a specified limit
limit = 1000000
primes = generate_primes(limit)

# Calculate the average of residues for each prime number using smaller primes
averages = []
for prime in primes:
    smaller_primes = [p for p in primes if p < prime]
    avg_residue = average_residues_prime_vs_prime(prime, smaller_primes)
    averages.append(avg_residue/prime)

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(primes, averages, marker='o')
plt.xlabel('Prime Numbers')
plt.ylabel('Average of Residues (Prime vs Prime)')
plt.title('Average of Residues for Prime Numbers vs Smaller Primes')
plt.grid(True)
plt.show()
```
